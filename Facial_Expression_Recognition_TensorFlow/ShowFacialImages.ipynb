```python
import numpy as np
import matplotlib.pyplot as plt
from Util import GetData
```

```python
labelMap = ['Anger', 'Disgust', 'Fear', 'Happy', 'Sad', 'Surprise', 'Neutral']

def main():
    X, Y = GetData(balanceOnes = False)
    
    while True:
        for i in range(7):
            x,y = X[Y==i], Y[Y==i]
            N = len(y)
            j = np.random.choice(N)
            plt.imshow(x[j].reshape(48,48), cmap='gray')
            plt.title(labelMap[y[j]])
            plt.show()
        prompt = input('Quit the program? Y or y:\n')
        if(prompt == "y"):
            break

if __name__ == '__main__':
    main()
```
